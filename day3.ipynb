```python
# 정렬 알고리즘
# list의 가지고 있는 요소들을 순서대로 열거하는 알고리즘

# 오름차순 정렬 
# 버블 정렬(Bubble Sort)
# 두 인접한 요소의 크기를 비교, 치환함으로써 정렬


array = [45, 25, 7, 82, 12]
print("정렬 전: ", array)

# step1

for i in range(4):
    if array[i] > array[i + 1]:
        temp = array[i]
        array[i] = array[i + 1]
        array[i + 1] = temp
print("정렬 후 : ", array)
for j in range(3):
    if array[j] > array[j + 1]:
        temp = array[j]
        array[j] = array[j + 1]
        array[j + 1] = temp
print("정렬 후 : ", array)

# step2
for m in range(4):
    for n in range(4 - m):
        if array[n] > array[n + 1]:
            temp = array[n]
            array[n] = array[n + 1]
            array[n + 1] = temp
print("정렬 후 : ", array)
```

```python
# 선택 정렬(Selection Sort)
# 가장 큰 값(혹은 가장 작은 값)의 위치를 찾아 교환함으로써 정렬(한 스텝 당 한 번만 일어남)

a = [98, 7, 70, 13, 25]
print("선택 정렬 전: ", a)

# 오름차순 정렬
#step1
max_index = 0 # 가장 큰 값의 위치 정보
for i in range(5):
    if a[i] > a[max_index]:
        max_index = i

temp = a[max_index]
a[max_index] = a[4]
a[4] = temp
print("선택 정렬 후: ", a)


for j in range(5):
    max_index = 0
    for i in range(5 - j):
        if a[i] > a[max_index]:
            max_index = i
    temp = a[max_index]
    a[max_index] = a[4 - j]
    a[4 - j] = temp
print("선택 정렬 후: ", a)
```

```python
# 탐색
# 원하는 요소나 값이 어느 위치에 있는가?
arr = [13, 35, 15, 11, 26, 72, 78, 13, 61, 90]
# 78은 어느위치에 있는지
search = 78
#Sepuential Search
# 모든 요소 비교
for i in range(len(arr)):
    if arr[i] == search:
        index = i
        break
print(f"{search}은 index : {index}위치에 있습니다")
```

```python
# 이진 탐색(Binary Search)
# 범위를 2분할하여 값을 탐색하는 기법

list4 = [1, 16, 30, 33, 66, 90, 7, 25, 41, 78]
list4.sort() # 선행 조건 : 오름차순, 내림차순 정렬!!
print(list4)
index = -1


# Logic
low_index = 0
high_index = 9

import time
search = 78

while low_index <= high_index:
        
    mid_index = (low_index + high_index) // 2
    if list4[mid_index] < search:
        low_index = mid_index + 1
    elif list4[mid_index] > search:
        high_index = mid_index - 1
    else:
        index = mid_index
        break

print(f"{search}는(은) index : {index}위치에 있습니다")
```

```python
# abs
def absolute(num):
    # reurn num if num > 0 else -num <- 가능
    if num < 0:
        return -num
    else:
        return num

print(absolute(-10)) # 10
print(absolute(10))  # 10
```

```python
# close10(num1, num2)
# 둘 중에 10에 더 가까운 수를 return하는
def close10(a, b):
    # if absolute(a - 10) <- a와 10 사이의 거리
    # if absolute(b - 10) <- b와 10 사이의 거리
    if absolute(a - 10) < absolute(b - 10):
        return a
    else:
        return b
    

num1 = int(input("첫 번째 정수 입력 : "))
num2 = int(input("두 번째 정수 입력 : "))
print(close10(num1, num2))
```

```python
list5 = [10, 5, 7, 20, 3]
# list의 가장 큰 값을 반환하는 
# 나만의 maximize() 정의
def maximize(list500):
    max_value = list500[0]
    for num in list500:
        if num > max_value:
            max_value = num

    return max_value
print(maximize(list5)) # 20
```
