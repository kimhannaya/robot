```python
# 나만의 어플리케이션 만들어 보자
from PyQt5.QtWidgets import QApplication, QWidget
import sys # 해당 App을 우리 윈도우에 띄우기 위해

# 처음부터 구현하지 않고 QWudge 상속 받아서 만들자
class MyApp(QWidget):
    def __init__(self):
        # 부모 안에 있는 field 가져다 쓰자
        super().__init__()
        self.initUI()

    # 초기 화면 구성
    def initUI(self):
        self.setWindowTitle("exam01")
        self.move(1000, 500)
        self.resize(400, 200) # width * height (pixel)
        self.show()

if __name__ == "__main__":
    # 내가 직접 실행했을 때만 실행되는 부분
    # 현재 동작하고 있는 App을 띄우기 위한 인스턴스가 있는지 여부
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv) # 윈도우 시스템에 띄우기 위해
    exam = MyApp()
    sys.exit(app.exec_()) # 윈도우 창이 꺼지면 app도 없애버리자
```

```python
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton
import sys 

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    # 초기 화면 구성
    def initUI(self):
        self.setWindowTitle("exam02")
        self.setGeometry(100, 500, 400, 200) # move, resize 한 번에
        button = QPushButton("눌러봐잇", self) # (문구, 부모)
        button.move(50, 70) # 왼쪽 상단 기준 얼마나 떨어져 있을지
        button.resize(100, 50)
        button.resize(button.sizeHint()) # 버튼 사이즈 알아서 추천

        # 버튼을 눌렀을 때 실행되는 함수 정의
        def set_color():
            self.setStyleSheet("background-color : #FF0000")
            button.setText("바보 ㅋㅋ")
        button.clicked.connect(set_color)

        # 버튼을 눌렀을 때 실행되는 함수 정의
        # 이렇게 정의 된 함수는 한 번만 쓰이니까
        # 정의하지 말고 바로 쓰자 -> lambda expression
        # df.apply(타이타닉 Sex Female -> 1, 0)
        button.pressed.connect(lambda : self.setStyleSheet("background-color: #F86796"))
        button.released.connect(lambda : self.setStyleSheet("background-color: #0000FF"))
        
        self.show()

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv) 
    exam = MyApp()
    sys.exit(app.exec_()) 
```

```python
from PyQt5.QtWidgets import (
                        QApplication, QWidget, QLineEdit, QPushButton, QLabel,
                        QRadioButton, QCheckBox, QVBoxLayout, QHBoxLayout
)
import sys

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    # 초기화면 구성
    def initUI(self):
        self.setWindowTitle("exam03")
        self.setGeometry(100, 500, 400, 200) # move, resize 한 번에
        
        qle_id = QLineEdit(self)
        qle_pw = QLineEdit(self)

        btn_login = QPushButton("로그인", self)

        qle_id.setGeometry(50, 50, 200, 25)
        qle_id.setPlaceholderText("ID를 입력하세요")
        qle_pw.setGeometry(50, 100, 200, 25)
        qle_pw.setPlaceholderText("PW를 입력하세요")
        qle_pw.setEchoMode(QLineEdit.Password)
        btn_login.setGeometry(100, 150, 100, 25)

        def login():
            # qle_id에 있는 문자열 정보 받아오고
            # qle_pw에 있는 문자열 가져오고
            login_id = qle_id.text()
            login_pw = qle_pw.text()
            if login_id == "hannaya" and login_pw == "1212":
                self.setStyleSheet("background-color: #FF0000")
            else:
                self.setStyleSheet("background-color: #0000FF")

        btn_login.clicked.connect(login)
        self.show()

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```

```python
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (
                        QApplication, QWidget, QLineEdit, QPushButton, QLabel,
                        QRadioButton, QCheckBox, QVBoxLayout, QHBoxLayout
)
from PyQt5.QtGui import QPixmap
import sys

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    # 초기화면 구성
    def initUI(self):
        self.setWindowTitle("exam04")
        self.setGeometry(800, 300, 400, 600) # move, resize 한 번에
        # 수직 Box
        #  - 위젯
        #  - 수평 Box
        #  - - 버튼1, 버튼2
        vbox = QVBoxLayout()
        vbox.addStretch(1)
        self.ql = QLabel("연습용", alignment=Qt.AlignCenter)

        # 이미지 로드
        self.idx = 0
        path = f"imgs/img{self.idx + 1}.jpeg"
        qpm = QPixmap(path)
        qpm = qpm.scaled(400, 400)
        self.ql.setPixmap(qpm)
        

        def next_img():
            self.idx += 1
            if self.idx >= 5:
                self.idx = 0
            path = f"imgs/img{self.idx + 1}.jpeg"
            qpm = QPixmap(path)
            qpm = qpm.scaled(400, 400)
            self.ql.setPixmap(qpm)
        def prev_img():
            self.idx -= 1
            if self.idx  < 0:
                self.idx = 4
            path = f"imgs/img{self.idx + 1}.jpeg"
            qpm = QPixmap(path)
            qpm = qpm.scaled(400, 400)
            self.ql.setPixmap(qpm)
        
        vbox.addWidget(self.ql)
        vbox.addStretch(1)

        hbox = QHBoxLayout()
        btn1 = QPushButton("<")
        btn1.setStyleSheet("background-color:#CC99FF")
        btn1.clicked.connect(prev_img)
        btn2 = QPushButton(">")
        btn2.setStyleSheet("background-color:#CC99FF")
        btn2.clicked.connect(next_img)
        hbox.addStretch(1)
        vbox.setSpacing(40)
        
        
        hbox.addWidget(btn1)
        hbox.addWidget(btn2)
        hbox.addStretch(1)

        vbox.addLayout(hbox)
        vbox.addStretch(1)

        self.setLayout(vbox)
        
        self.show()

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```
