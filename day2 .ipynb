```python
# 반복문

# while : 대게 반복 횟수가 정해져 있지 않을 때
# for   : 대게 반복 횟수가 정해져 있을 때

# while문

num = int(input("자연수를 입력하세요: "))
total = 0

while num > 0:
    total += num
    print(f"누적합 : {total}")
    num = int(input("자연수를 입력하세요: "))
```

```python
#break 키워드
# 가장 가까운 반복문을 탈출함
# 조건식 자리에  True
# 무한 반복문 : 탈출 조건 명시
total = 0
while True:
    num = int(input("자연수를 입력하세요 : "))
    if num > 0:
        break
print(f"누적 합 : {total}")
```
